```python
# Import Dependencies
from ultralytics import YOLO
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')
```

```python
# Load a Pretrained YOLOv10 Model
# Currently using v8s because we're running this on a raspberry pi 3b+
model = YOLO('/content/drive/MyDrive/cocomaturity/checkpoints/train4/weights/last.pt')
```

```python
# Import Data from Google Drive
!cp -r /content/drive/MyDrive/cocomaturity /content/

# Prepare Dataset for Training
!unzip /content/cocomaturity/CocoMaturityDetect.v2-3101.yolov9.zip -d /content/cocomaturity/

# Update data.yaml to use Collab Path
!sed -i 's|\.\.|/content/cocomaturity|g' /content/cocomaturity/data.yaml
!cat /content/cocomaturity/data.yaml
```

```python
# Train the model initially
checkpoint_path = '/content/drive/MyDrive/cocomaturity/checkpoints'
results = model.train(data='/content/cocomaturity/data.yaml', epochs=100, imgsz=640, batch=42, project=checkpoint_path, verbose=True, resume=True)

```
